```python
def gets_info():
    links = []
    t= []
    title= []
    author = []
    
    for i in range(1):
        url = ('https://newsapi.org/v2/everything?'
               'q=bitcoin&'
               'from=2017-11-08&'
               'to=2018-01-08&'
               'language=en&'
               'sortBy=publishedAt&'
               'page={}&'
               'pageSize=5&'
               'apiKey=8ba091b7a47b4c9a9162a83ca72eb1ca'
               ).format(i+1)
        everything = requests.get(url).json()['articles']
        
        links.extend([i['url'] for i in everything])
        t.extend([i['publishedAt'] for i in everything])
        title.extend([i['title'] for i in everything])
        author.extend([i['author'] for i in everything])
        time.sleep(randint(5, 10))
    df = pd.DataFrame({'links': links,
                      'time': t,
                      'title': title,
                      'author': author})
    
    return df
        
```

```python
def get_article(http):
    src = requests.get(http).content
    sp = bs.BeautifulSoup(src, 'lxml')
    cnt = sp.find_all('p')
    a=[d.text for d in cnt]
    text = " __ ".join(a)
    return text
```

```python
def text(links):
    
    text = [get_article(l) for l in links]
    
    return text
```
