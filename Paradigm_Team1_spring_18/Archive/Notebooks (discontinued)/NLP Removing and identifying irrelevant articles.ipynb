```python
#BY CONTENTS
#Check how many articles are relevant to Bitcoin/crypto, and track indices of articles that are not

keywords = ["Bitcoin", "bitcoin", "Cryptocurrency", "cryptocurrency"] #Keywords to look for - This can be expanded
irrelevant_article_indices = [] #To keep track of indices of articles with no keywords

for article_index in range(0, len(short.index)):
    article = short['contents'][article_index] #Grab the contents of the current article
    num_keywords = 0 #Initialize keyword counter
    
    for keyword in keywords: #This loop counts the number of keywords from the list that appear in the article
        if keyword in article:
            num_keywords += 1 
        
    if num_keywords == 0: #If there are no keywords in the article, indicate that and track the index
        print("Did not find any keywords in article", article_index)
        irrelevant_article_indices.append(article_index)
        
    

print("Out of", len(short.index), "articles,", 
      len(irrelevant_article_indices), "have nothing to do with Bitcoin or cryptocurrency.")
```

```python
#Remove irrelevant articles from master CSV

for index in irrelevant_article_indices:
    short = short.drop([index], axis=0)
    
short = short.reset_index()
short = short.drop(['index'], axis = 1)

short.head()
```

```python
#reading the sentiments
sentiments = pd.read_table('vader_lexicon.txt', names=['token', 'polarity','SD','human ratings'])
sentiments.tail(30)
```

```python
sent = sentiments.drop(['SD','human ratings'], axis=1).set_index('token')
sent
```

```python
# Save your regex in punct_re- getting rid of punctuation 
punct_re = r'[^\w\s]'
short['no_punc'] = short['contents'].str.lower().apply(lambda x: re.sub(punct_re,' ', x))
short.head()
```

```python
#splitting into separate words for sentiment analysis
short['split']= short.no_punc.str.split()
tidy_short = (pd.DataFrame(short.split.apply(pd.Series).stack())
               .reset_index(level=1)
               .rename(columns={'level_1':'num', 0: 'word'}))
short=short.drop('split', axis=1)
```

```python
tidy_short.head()
```

```python
tidy_short['id'] = tidy_short.index
tidy_short.head()
```

```python
tidy_sort = tidy_short.merge(sent, how='left', left_on='word', right_index=True)
```

```python
#merging polarities with the words we have
polarities = (tidy_short
              .merge(sent, how='left', left_on='word', right_index=True)
              .fillna(value=0)
              .groupby('id')
              .sum()
              .drop('num', axis=1))
polarities.tail()
```

```python
#putting back the polarities with the dataframe we have 
senti = short.merge(polarities, how='left', left_index=True, right_index=True)
senti
```
